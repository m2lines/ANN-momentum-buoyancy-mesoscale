```python
def plot_density(ds, title='No param'):
    plt.figure(figsize=(8,3))

    plt.subplot(1,2,1)
    ds.rhoinsitu.isel(Time=0, yh=0).plot.contourf()
    plt.gca().invert_yaxis()
    plt.title('')

    plt.subplot(1,2,2)
    ds.rhoinsitu.isel(Time=-1, yh=0).plot.contourf()
    plt.gca().invert_yaxis()
    plt.title('')

    plt.suptitle(title)

    plt.tight_layout()
```

```python
plot_density(ds)
plot_density(ds_GM1000, title='GM 1000')
plot_density(ds_ANNGM1000, title='ANN GM 1000')
```

```python
plt.figure(figsize=(8,3))

plt.subplot(1,2,1)
idT = 0
(ds_GM1000.rhoinsitu - ds_ANNGM1000.rhoinsitu).isel(Time=idT, yh=0).plot.contourf()
plt.title('delta rho GM1000 - ANN GM1000')

plt.subplot(1,2,2)
idT = -1
(ds_GM1000.rhoinsitu - ds_ANNGM1000.rhoinsitu).isel(Time=idT, yh=0).plot.contourf()
plt.title('delta rho GM1000 - ANN GM1000')

plt.tight_layout()
```
